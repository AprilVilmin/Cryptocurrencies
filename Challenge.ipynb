```python
import pandas as pd
```

```python
file_path = "crypto_data.csv"
crypto_df = pd.read_csv("crypto_data.csv")
crypto_df.head()
```

```python
#https://practicaldatascience.co.uk/data-science/how-to-select-filter-and-subset-data-in-pandas-dataframes
df_crypto = crypto_df[crypto_df['IsTrading']=='True']
df_crypto.head()
```

```python
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html
no_IsTrading_df = df_crypto.drop('IsTrading', axis=1)
no_IsTrading_df
```

```python
#https://www.machinelearningplus.com/pandas/pandas-dropna-how-to-drop-missing-values/#:~:text=in%20specific%20columns-,DataFrame.,names%20to%20the%20subset%20parameter.
no_nulls_df = no_IsTrading_df.dropna(inplace= False, axis=0)
no_nulls_df
```

```python
df_coins_mined = no_nulls_df[no_nulls_df['TotalCoinsMined'] > 0 ]
df_coins_mined.head()
```

```python
#https://stackoverflow.com/questions/34682828/extracting-specific-selected-columns-to-new-dataframe-as-a-copy
df_coin_name = pd.DataFrame(df_coins_mined["CoinName"], index = df_coins_mined.index)
df_coin_name.head()
```
